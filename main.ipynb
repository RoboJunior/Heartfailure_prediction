```python
import pandas as pd
df = pd.read_csv("heart.csv")
df
```

ðŸŸ¢ Removing outliers using zscore

```python
df["resting_bp_zscore"] = (df.RestingBP - df.RestingBP.mean())/ df.RestingBP.std()
```

```python
df[df["resting_bp_zscore"]>3]
```

```python
df[df['resting_bp_zscore']<-3]
```

```python
df_new = df[(df.resting_bp_zscore>-3)&(df.resting_bp_zscore<3)]
```

```python
df_new
```

```python
df_new['Cholesterol_zscore'] = (df_new.Cholesterol - df_new.Cholesterol.mean())/df_new.Cholesterol.std()
```

```python
df_new
```

```python
df_new1 = df_new[(df_new.Cholesterol_zscore>-3)&(df_new.Cholesterol_zscore<3)]
```

```python
df_new1
```

```python
df_new1['MaxHR_zscore'] = (df_new1.MaxHR - df_new1.MaxHR.mean())/df_new1.MaxHR.std()
```

```python
df_new1
```

```python
df_new2 = df_new1[(df_new1.MaxHR_zscore>-3)&(df_new1.MaxHR_zscore<3)]
```

```python
df_new2
```

```python
df_new2['Oldpeak_zscore'] = (df_new2.Oldpeak - df_new2.Oldpeak.mean()) / df_new2.Oldpeak.std()
```

```python
df_new2
```

```python
final_df = df_new2[(df_new2.Oldpeak_zscore>-3)&(df_new2.Oldpeak_zscore<3)]
```

```python
final_df
```

ðŸŸ¢ After clearing the outliers here we can see clearly the dataset has reduced from 918 to 899

ðŸŸ¢ And We are removing the columns which are not needed for data modeling

```python
final_df.drop(["resting_bp_zscore","Cholesterol_zscore","MaxHR_zscore","Oldpeak_zscore"],axis=1,inplace=True)
```

Here we got the dataframe we all the columns we need

```python
final_df
```

ðŸŸ¢ Converting the text columns with one hot encoding

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
final_df['Sex'] = label_encoder.fit_transform(final_df['Sex'])
```

```python
final_df
```

```python
final_df['ChestPainType'] = label_encoder.fit_transform(final_df['ChestPainType'])
final_df['RestingECG'] = label_encoder.fit_transform(final_df['RestingECG'])
final_df['ExerciseAngina'] = label_encoder.fit_transform(final_df['ExerciseAngina'])
final_df['ST_Slope'] = label_encoder.fit_transform(final_df['ST_Slope'])
```

```python
final_df
```

ðŸŸ¢ Here you can see the dataset after onehot encoding process

```python
final_df.describe()
```

ðŸŸ¢ Applying scaling to the model before training the model

```python
X = final_df.drop("HeartDisease",axis=1)
y = final_df['HeartDisease']
```

```python
X,y
```

ðŸŸ¢ Here we are using StandardScaler from sklearn for getting the standard deviation of all the features

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

ðŸŸ¢ Here we are going to build a classification model , Here are chosing the Logistic Regression model for constructing the model

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X_scaled,y,test_size=0.2,random_state=30)
```

```python
from sklearn.linear_model import LogisticRegression

logistic_regression = LogisticRegression()
logistic_regression.fit(X_train,y_train)
logistic_regression.score(X_test,y_test)
```

ðŸŸ¢ Saving the model as the model has the best performance before pca 

ðŸŸ¢ The Model performance after PCA was worse so i am using this model for predictions

```python
import pickle
filename = 'final_model.sav'
saved_model = pickle.dump(logistic_regression,open(filename,'wb'))
```

```python
import pickle

loaded_model = pickle.load(open('final_model.sav','rb'))
result = loaded_model.score(X_test,y_test)
result
```

```python
y_pred = loaded_model.predict(X_test)
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(loaded_model.score(X_test, y_test)))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

ðŸŸ¢ Here we are using PCA for dimension reduction

```python
from sklearn.decomposition import PCA

pca = PCA(0.95)

X_pca=pca.fit_transform(X)
X_pca.shape
```

```python
pca.explained_variance_ratio_
```

```python
pca.n_components_
```

```python
X_train_pca,X_test_pca,y_train,y_test = train_test_split(X_pca,y,test_size=0.2,random_state=30)
```

```python
model2 = LogisticRegression()
model2.fit(X_train_pca,y_train)
model2.score(X_test_pca,y_test)
```

ðŸŸ¢ Here you can see the model performance is reduced while we are applying PCA to the dataset

ðŸŸ¢ Accuracy of the model without PCA:83
ðŸŸ¢ Accuracy of the model with PCA:66

ðŸŸ¢ So using PCA in this dataset is not recommended
